```python
# 3. Describe three different data types.

#Ans:
"""
Brodly, there are three different data types. They are:
1) Numeric
2) Sequence
3) Others

1) Numeric:
    This type of data type have numeric value. It is further categorized as:
    i) Integer - This data type includes whole numbers (both positive and negative) and zero. And it doesn't 
                 include fraction or decimal. The limit of integer data type is not defined in Python.
                 Eg: 45, 23507235902, 203832095802349623974634975273
                 
    ii) Float -  This data type includes all real numbers and associated with decimal even if the number is
                 whole number. Eg: 7.0, 32.6235223465, -32.09237502952666664357
                 
    iii) Complex - This data type includes complex numbers. A complex number has two component i.e. real part
                   and imaginary part. Real part denotes the real number and Imaginary part is the imaginary
                   number which doesn't exist in reality but helps in solving polynomial problems and is
                   suffixed with notation "j". This "j" denotes square root of (-1). 
                   Eg: 7+8j, -3-9j
                   
2) Sequence:
    This is further categorized into 2 types:
    i) Immutable - The value of the item cannot be changed at any index.
                   Eg: String and Tuple datatype
                   String - It is a collection of one or more characters or integers or operators or 
                            expressions, enclosed within single or double quotes. 
                            Eg: "1233", "Abc", "+*", "4+7"
                   Tuple - It is a ordered collection of items which can include different data types including
                           tuple itself but seperated by commas and enclosed with normal parenthesis i.e. ().
                           Eg.: (1, 'xyz', [2,3,5], (9, 73, 8), {3, 5, 9} )
    ii) Mutable - The value of the item can be changed at any index.
                  Eg.: List datatype
                  List - It is also ordered collection of items which can include different data types as
                         Tuple. Items are seperated by commas and enclosed with square parenthese i.e. [].
                         Items inside List can be List can be changed but not in Tuple.
                         Eg.: [25, 'abc', (3, 'abc', True), [32, 'Y', 8+8j]]
    
3) Others:
    Others include:
    i) Sets - It is a set unordered collection of "unique" items of different data types, which is not iterable
              and not mutable. The items in Set have no duplicate values and are enclosed in curly parenthesis 
              i.e. {}.
              Note: List, Dictionary and set itself cannot be part of collection of set
              Eg.: {23, 'abc', (3,4)}
    ii) Dictionary - It is set of unordered collection of items in key-value pair which are seperated with 
                     colon symbol and items are sepearted by commas and also enclosed with curly parenthesis
                     as Set i.e. {}.
                     Eg.: {'Key1' : 'Value', 'Key2' : 352}

"""
```

```python
#4. What is an expression made up of? What do all expressions do?

#Ans:
"""
An Expression is made of two (or more) operand and one (or more) numeric operator respectively, to execute to 
get a intended outcome. 
An operand can be value or variable. Minimum 2 operands are required to perform an operation and both must be
same data type.
An operator are symbols to perform numeric and logical operation

Eg.: a = 5, b = 8. a + b is a correct expression.
     a = 'v', b = 9. a + b is an incorrect expression, as they are of different datatypes.
     
     
An Expression always perform operation to produce a result.

"""
```
