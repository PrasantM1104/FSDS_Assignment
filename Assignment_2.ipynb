```python
#2. What are the three different types of Boolean operators?

#Ans:
"""
The three different types of Boolean opeartors are:
'And', 'Or' and 'Not'

'And' operator returns True value only when 2 or more conditions are met or are True.

'Or' operator returns True value when either of the condition is met or is True.  
    
'Not' operator returns True if condition is False and vice versa.
"""
```

```python
#3. Make a list of each Boolean operator's truth tables (i.e. every possible combination of Boolean values 
#   for the operator and what it evaluate ).

#Ans:
"""
The truth table of Boolean operator 'And' is:
    False and False --> False
    True  and False --> False
    False and True  --> False
    True  and True  --> True
'And' operator returns True value only when 2 or more conditions are met or are True.
Eg. 4 > 5 and 100 < 50   --> False
    4 < 5 and 100 < 50   --> False
    4 > 5 and 100 > 50   --> False
    4 < 5 and 100 > 50   --> True
    
The truth table of Boolean operator 'Or' is:
    False or False --> False
    True  or False --> True
    False or True  --> True
    True  or True  --> True
'Or' operator returns True value when either of the condition is met or is True.
Eg. 4 > 5 or 100 < 50   --> False
    4 < 5 or 100 < 50   --> True
    4 > 5 or 100 > 50   --> True
    4 < 5 or 100 > 50   --> True
    
The truth table of Boolean operator 'Not' is:  
    True  --> False
    False --> True
'Not' operator returns True if condition is False and vice versa.
Eg. not(4 > 5)  --> True
    not(4 < 5)  --> False
"""
```
